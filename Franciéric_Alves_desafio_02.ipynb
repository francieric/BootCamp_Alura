# Por que os brasileiros estão deixando de se vacinar?

<div align="justify"> 

Em queda há **cinco anos**, as coberturas vacinais do Programa Nacional de Imunizações (PNI) não atingem nenhuma meta no calendário infantil desde 2018, dados estes apresentados no início de outubro na Jornada Nacional de Imunizações.  

O Programa Nacional de Imunização (PNI)
é referência mundial. O Brasil foi pioneiro na incorporação de diversas vacinas no calendário do Sistema Único do Saúde (SUS) e é
um dos poucos países no mundo que ofertam
de maneira universal um rol extenso e abrangente de imunobiológicos. Porém, a alta taxa
de cobertura, que sempre foi sua principal
característica, vem caindo nos últimos anos,
conforme demonstra o quadro na página ao
lado, colocando em alerta especialistas e profissionais da área.

Tido como modelo mundial de política pública eficaz nas últimas décadas, o Programa Nacional de Imunizações (PNI) está a perigo. Uma série de fatores levaram o país aos níveis mais baixos de cobertura dos últimos 24 anos.

As últimas metas de imunização para o público infantil atingidas no país, em 2018, foram de 99,72% do público-alvo para a BCG, e de 91,33% para o da vacina contra o rotavírus humano. Para ambas, a meta é superar os 90%, patamar que não foi atingido em 2019, apesar de terem continuado acima dos 80%. Já até 2 de outubro de 2020, a taxa de imunização do público-alvo da BCG chegou a 63,88%, e a vacina contra o rotavírus, a 68,46%.

A maior cobertura atingida no calendário infantil até outubro de 2020 foi na vacina Pneumocócica, com 71,98%. No ano passado, essa mesma vacina chegou a 88,59% do público-alvo. Entre as 15 vacinas do calendário infantil, o que inclui a segunda dose da Tríplice Viral, metade não bate as metas desde 2015, o que inclui a vacina contra poliomielite.

A queda nas coberturas vacinais durante a pandemia foi um fenômeno sentido globalmente. Segundo a Organização Mundial da Saúde (OMS), 125 campanhas de vacinação que estavam marcadas para o primeiro semestre de 2020 foram adiadas. O problema da interrupção dos serviços de vacinação levou a OMS e o Fundo das Nações Unidas para a Infância (Unicef) a alertarem que 117 milhões de crianças em 37 países poderiam deixar de receber a vacina contra o sarampo, que também provocou surtos em diversas partes do mundo nos últimos anos, incluindo o Brasil.

A queda nas coberturas desafia o Programa Nacional de Imunizações do Sistema Único de Saúde, considerado um dos mais amplos e bem sucedidos do mundo. O programa teve um aumento expressivo nas taxas de vacinação entre 1980 e 1995, ano a partir do qual as taxas ficaram estáveis em patamares elevados, e, em alguns casos, superiores a 100%. **O recuo teve início em 2015, e, antes da pandemia, já pesavam fatores como horários de funcionamento das unidades de saúde, a circulação de informações falsas sobre a segurança das vacinas e até mesmo a impressão de que as doenças imunopreveníveis já deixaram de existir**.

Surge uma ***HIPÓTESE***:

**Importando e ajustando o Dataset**

```
#Importando o Dataset removendo os dados irrelevantes
dados = pd.read_csv("/content/drive/MyDrive/dataset_desafio02_PNI.csv",
                    encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=20,
                    thousands=".", decimal=",")
dados.tail()
```

**Analisando a estrutura do Dataframe**

```
dados.info()
```

```
dados.shape
```

```
#renomenando a coluna
dados.rename(columns={'Unidade da Federação':'uf'}, inplace=True)
dados.tail()
```

```
# alterando o indice
dados.set_index('uf',inplace=True)
dados.tail()
```

**Adicionando e ajustando a coluna região ao DataFrame**

```
#Adicionando as regiões ao Df
dados["regiao"] = dados.index.str[0]
regioes_map = {1:"Norte",2:"Nordeste",3:"Sudeste",4:"Sul",5:"Centro-Oeste"}
dados["regiao"] = dados["regiao"].map(lambda x: regioes_map[int(x)])
dados["regiao"]
```

```
#removendo os valores numericos antes do nome da UF
dados.index= dados.index.str[3:]
dados.tail()
```

```
#Removendo a coluna total
dados.drop('Total',axis=1,inplace=True)
dados.tail()
```

**Selecionando o intervalo de estudo - 2015 a 2019**

```
#Fazendo o Slice no período de estudo desejado.
dados[dados.columns[-6:]]
```

**Fazendo uma cópia do Dataframe e resetando  o index.**

```
dados_abertos = dados.copy()
dados_abertos.reset_index(inplace=True)
dados_abertos.head()
```

**Mudando a disposição dos dados na tabela utilizando MELT**

```
dados_abertos = dados_abertos.melt(id_vars=['regiao','uf'],var_name = 'ano', value_name = 'indice_PNI')
dados_abertos
```

**Agrupando os dados por Regiao, UF e Ano.**

```
#AGRUPANDO OS DADOS POR REGIAO, UF e ANO.
regioes_list = ['Sudeste','Nordeste','Sul','Norte','Centro-Oeste']
por_ano_estados_selecionados = dados_abertos.query("regiao== @regioes_list and ano>='2015'").groupby(['regiao','uf','ano']).sum().reset_index()
por_ano_estados_selecionados.head()
```

##**Analise do Percentual de Cobertura de Vacinas por Região - 2015 a 2019**

```
por_ano_estados_selecionados_2015=por_ano_estados_selecionados.query("ano=='2015'").reset_index()
por_ano_estados_selecionados_2016=por_ano_estados_selecionados.query("ano=='2016'").reset_index()
por_ano_estados_selecionados_2017=por_ano_estados_selecionados.query("ano=='2017'").reset_index()
por_ano_estados_selecionados_2018=por_ano_estados_selecionados.query("ano=='2018'").reset_index()
por_ano_estados_selecionados_2019=por_ano_estados_selecionados.query("ano=='2019'").reset_index()
```

```
por_ano_estados_selecionados_2016['variacao_ano_a_ano']=por_ano_estados_selecionados_2016['indice_PNI']-por_ano_estados_selecionados_2015['indice_PNI']
por_ano_estados_selecionados_2016.sort_values(['variacao_ano_a_ano'])
```

```
por_ano_estados_selecionados_2017['variacao_ano_a_ano']=por_ano_estados_selecionados_2017['indice_PNI']-por_ano_estados_selecionados_2016['indice_PNI']
por_ano_estados_selecionados_2017.sort_values(['variacao_ano_a_ano'],ascending=False)
```

```
por_ano_estados_selecionados_2018['variacao_ano_a_ano']=por_ano_estados_selecionados_2018['indice_PNI']-por_ano_estados_selecionados_2017['indice_PNI']
por_ano_estados_selecionados_2018.sort_values(['variacao_ano_a_ano'])
```

```
por_ano_estados_selecionados_2019['variacao_ano_a_ano']=por_ano_estados_selecionados_2019['indice_PNI']-por_ano_estados_selecionados_2018['indice_PNI']
por_ano_estados_selecionados_2019.sort_values(['variacao_ano_a_ano'])
```

**Gráfico da Cobertura Vacinal por Região entre 2015 e 2019**

```
#PLOTAGEM VISUAL DE TODAS AS REGIOES - IDENTIFICACAO DE ALGUM INDICIO DE QUE REGIAO PODERIA TER MAIOR PARTICIPAÇÃO NA QUEDA DOS INDICES
ax=sns.catplot(data =por_ano_estados_selecionados, x = 'ano', y='indice_PNI',kind='point',col='regiao',hue='uf',palette = 'bright',legend=False)
ax.set_ylabels('Percentual da Meta Atingido')
ax.fig.set_size_inches(20,7)
plt.xticks(rotation= 0)
plt.legend(title="UF", bbox_to_anchor=[1.5, 0.8], loc='upper right')
plt.show()
```

Conforme apresentado este estudo tem como objetivo analisar o percentual de Cobertura Vacinal do Brasil no período de 2015 a 2019 (pré-pandemia), aprofundando essa analise por Regiões e Estados, verificando a existência de algum aspecto pontual regional ou Estadual que possa ter impactado na queda do Índice de Cobertura Vacinal nos últimos 05 anos. Essa analise seguiu no sentido de
Identificar potenciais Regiões e/ou Estados cujo comportamento quanto a Cobertura Vacinal indique a adoção de maiores investimentos governamentais para esclarecer a população, visto que, para alguns autores, a circulação de informações falsas sobre a segurança das vacinas e até mesmo a impressão de que as doenças imunopreveníveis já deixaram de existir são fatores que estão afetando a adesão aos programas de vacinação, logo, surge a hipótese de que se tivermos  investimentos mais direcionados a conscientização da população, em especial nas Regiões e Estados que tiveram uma maior queda na cobertura vacinal nos últimos anos, poderíamos ter uma ação que amenizaria ou até resolveria esse problema.

Analisando os gráficos por Região nos faz perceber que a queda do percentual de cobertura vacinal oscilou muito ao longo do período de 2015 a 2019, considerando a variação entre um ano e outro (exemplo: variação de 2016 em relação a 2015), podemos perceber que a Região Sul apresentou as menores quedas e manteve seus índices de cobertura vacinal sempre entre os mais elevados e em geral as Regiões Nordeste e Norte indicariam que carecem de mais investimento, sendo que nos últimos 03 anos (2017, 2018 e 2019), em média, a Região Sul apresentou os melhores percentuais de cobertura vacinal do Brasil. Enquanto nos anos de 2016, 2017, 2018 e 2019, em média, a Região Nordeste apresentou os piores percentuais do Brasil.

Outra analise feita foi verificando a “queda” desse percentual de cobertura vacinal de 2015 a 2019, ou seja, como estavam os índices em 2015 e como fecharam em 2019. Esta analise evidenciou que na Região Centro-Oeste por mais que Mato Grosso do Sul, no período de 2015 a 2019, tenha tido seu percentual de cobertura vacinal diminuído em 60,86% este percentual ainda é o maior da Região Centro-Oeste. 
O mesmo acontece na Região Nordeste, onde o Estado do Ceará teve sua cobertura vacinal diminuída em 60,18%, mas ainda é o Estado com maior índice da Região Nordeste com 47,53%. Na região Norte o Estado de Rondônia teve sua cobertura vacinal diminuída em 59,83% e ocupa a segunda colocação na Região em cobertura com 51,44%. Na Região Sudeste o Rio de Janeiro apresentou a maior queda (70,15%) e apresenta o menor índice de cobertura vacinal (25,50%) entre todos os Estados da Federação. Na Região Sul o Estado de Santa Catarina apresentou uma queda de 40,10% em seu índice, mas ainda é o Estado com o maior percentual de cobertura vacinal na Região Sul com 60,59%, sendo o Estado da Federação com o maior percentual. 

Com base na analise feita é possível criar um ranking, em ordem crescente de prioridade, de investimento em ações esclarecedoras dos principais pontos a respeito do Programa Nacional de Imunização (PNI) por região para melhoria do percentual de cobertura vacinal::

1.   Nordeste
2.   Norte
3.   Sudeste
4.   Centro-Oeste
5.   Sul

Outra ação pontual seria investigar e direcionar esforços aos Estados que tiveram as maiores baixas nos percentuais de cobertura da vacinação no período de 2015 a 2019, por exemplo os casos dos Estados que tiveram pelo menos 50% de queda no seu índice de cobertura vacinal: Rio de Janeiro (70,15%), Mato Grosso do Sul (60,86%), Ceará (60,18%), Rondônia (59,83%), Roraima (57,23%), Pernambuco (56,18%), Bahia (54,38%), Maranhão (54,35%), Mato Grosso (53,94%), São Paulo (52,45%) e Rio Grande do Norte (51,07%).
